#Using these things

Next, we want to apply the functions that we have just been experimenting with to the aggregated data set.  We join all the sheets that we stored in df_list.  Then we extract rows that mention 'Teacher Education - Mathematics'.  We store this as 'allm'.

```
allsub=pd.concat(df_list)

allm = allsub.loc[allsub.Category == "Teacher Education - Mathematics"]

allm
```

Next, we aggregate the data, as we did above.  Now, each state has data from numerous years.

```
allmgr = allm[['State','ReportYear', 'Prepared']].groupby(['State','ReportYear']).sum()

allmgr

```

Finally, we change the structure of the frame using .umstack(), so we have one row for each state and one column for each year.  We wrap 
pd.DataFrame(_____, dtype=object)
around
allmgr.unstack()
to avoid the data from being converted to floating point.

```
pd.DataFrame(allmgr.unstack(), dtype=object)
```

An alternate approach is to make a table rather that a multi-indexed object.

```
allmgrdan = allm[['State','ReportYear', 'Prepared']].groupby(['State','ReportYear'],as_index = False).sum()

allmgrdan
```

```
allmgrdan.pivot(index='State', columns='ReportYear', values='Prepared')
```

#We now put this all together as a function that we can apply to any subject.

```
df_list = []
for x in list(range(2012, 2019)):  
    df_list.append(pd.read_excel("AllStates-{}.xls".format(x), 'PreparedBySubject'))
all = pd.concat(df_list)

def display(subject):
  allthis = all.loc[all.Category == subject]
  allthis = allthis[['State','ReportYear', 'Prepared']]
  allthis = allthis.groupby(['State','ReportYear']).sum().unstack()
  return pd.DataFrame(allthis, dtype=object)
```
